# Cage Energy Minimization in CPP

This notebook calculates minimum energy configurations for A/B hDP mixes in tetrahedral cages.

**Key Mechanics**:
- Energy U from Coulomb + confinement.
- Mixes: 0A/4B to 4A/0B.
- Boltzmann population for stability.

Updated: January 15, 2026

```python
import numpy as np
import matplotlib.pyplot as plt

def cage_energy_minimization(n_A, n_B, cage_type='tetrahedral'):
    if cage_type == 'tetrahedral':
        vertices = 4
        bonds = 6
        r_min = 2.61e-35
        alpha = 7.3e-3
        E_central = -alpha / r_min
        E_hDP_A = 100e6 * 1.6e-19
        E_hDP_B = 120e6 * 1.6e-19
        E_total = E_central + n_A * E_hDP_A + n_B * E_hDP_B
        if abs(n_A - n_B) > 1:
            E_penalty = 10e6 * 1.6e-19 * (n_A - n_B)**2
            E_total += E_penalty
        return -E_total
    else:
        raise NotImplementedError(f"Cage type {cage_type} not implemented")

# Calculate for all mixes
mixes = []
energies = []
for n_A in range(5):
    n_B = 4 - n_A
    mixes.append(f"{n_A}A/{n_B}B")
    energies.append(cage_energy_minimization(n_A, n_B))

# Plot
plt.bar(mixes, energies)
plt.xlabel('Composition')
plt.ylabel('Binding Energy (arb. units)')
plt.title('Cage Energy Minimization')
plt.show()
```

## Conclusions
- 2A/2B minimum as expected.
- Imbalanced have penalty ~10–20%.
