# hDP Locus in Cages

This notebook simulates the arced locus of extra hDP oscillation in strange quark cages.

**Key Mechanics**:
- Vertex forces arc path.
- Energy contours show min U face.
- Variable from mix/history.

Updated: January 15, 2026

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Tetra vertex positions (simplified 2D projection)
vertices = np.array([[0,0], [1,0], [0.5, np.sqrt(3)/2], [0.5, 0.5]])  # Example
charges = [1, -1, 1, -1]  # Example mix

def energy_at_position(pos, central_charge=-1):
    U = 0
    for v, q in zip(vertices, charges):
        r = np.linalg.norm(pos - v)
        U += q / r if r > 0 else 0
    return U + central_charge / np.linalg.norm(pos)

# Minimize for equilibrium
result = minimize(energy_at_position, [0.5, 0.3])
print("Equilibrium position:", result.x)

# Plot energy contour
X, Y = np.meshgrid(np.linspace(0,1,100), np.linspace(0,1,100))
Z = np.array([energy_at_position([x,y]) for x,y in zip(X.ravel(), Y.ravel())]).reshape(X.shape)
plt.contourf(X, Y, Z)
plt.colorbar()
plt.title('Energy Contours in Tetra Cage')
plt.show()
```

## Conclusions
- Locus arcs to min U face.
- Variable based on charge mix.
- Persists until activation energy switch.
