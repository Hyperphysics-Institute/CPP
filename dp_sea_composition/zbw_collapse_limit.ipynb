# ZBW Collapse Limit in Black Hole Densities

This notebook models ZBW oscillation collapse when inner/outer SSV equal in black holes.

**Key Mechanics**:
- SSV vs. radius plot.
- Zeno convergence at equality.
- Limit when r_min = 0.

Updated: January 15, 2026

```python
import numpy as np
import matplotlib.pyplot as plt

# SSV model (simplified 1/r^2)
r = np.logspace(-35, -30, 100)  # Planck to black hole scales
ssv_inner = 1 / r**2
ssv_outer = 1 / (r + 1e-32)**2  # Slight offset for Zeno

# Plot
plt.plot(r, ssv_inner, label='Inner SSV')
plt.plot(r, ssv_outer, label='Outer SSV')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Radius (m)')
plt.ylabel('SSV (arb. units)')
plt.title('SSV Equality at Black Hole Densities')
plt.legend()
plt.show()

# Convergence point
delta_ssv = ssv_inner - ssv_outer
convergence_r = r[np.argmin(abs(delta_ssv))]
print("Convergence radius:", convergence_r)
```

## Conclusions
- Equality at ~10^{-32} m (black hole scale).
- ZBW neutralizes, allowing collapse.
- Zeno steps converge asymptotically.
