# Electron Binding Energy Calculation (Paper 1 Section 8)

Computes E_binding = 2 lattice units for minimal tetrahedral cage, then scales SSV_0 to match 0.511 MeV.

```python
# Parameters
num_shell_cps = 4
ssv_0_lattice = 1.0  # lattice units
m_e_mev = 0.511

# Potential from one shell CP
phi_shell = - ssv_0_lattice / 1.0  # d_0 = 1
phi_total = num_shell_cps * phi_shell

# Binding energy
e_binding_lattice = -0.5 * phi_total * ssv_0_lattice
print(f"E_binding (lattice units): {e_binding_lattice:.1f}")

# Scale SSV_0 to match physical units
ssv_0_mev = m_e_mev / e_binding_lattice
print(f"Scaled SSV_0: {ssv_0_mev:.4f} MeV")
print(f"Matches electron mass: {e_binding_lattice * ssv_0_mev:.3f} MeV")
```
