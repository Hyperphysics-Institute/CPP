# ZBW Energy Examples

Calculates E_ZBW = ½ m (c / r_eff)² · σ for different modes.

```python
import numpy as np

c = 3e8  # m/s
hbar = 1.0545718e-34  # J s
m_e = 9.1093837e-31  # kg
r_compton = hbar / (m_e * c)  # ≈ 2.426e-12 m
phi = (1 + np.sqrt(5)) / 2
sigma = lambda d: 120 ** (-d)

# Electron orbital (d=0, r_eff ≈ r_compton / phi)
r_eff_e = r_compton / phi
v_rel = 0.999 * c  # near-relativistic
E_orbital_e = 0.5 * m_e * v_rel**2 * sigma(0)
print(f"Electron orbital E_ZBW (approx): {E_orbital_e * 6.241509e18:.3f} eV")  # Convert J to eV

# Neutrino unbound (d=3, tiny v)
E_unbound_nu = 0.5 * m_e * (1e-3 * c)**2 * sigma(3)  # Example tiny v
print(f"Neutrino unbound example: {E_unbound_nu * 6.241509e18:.3e} eV")
```
