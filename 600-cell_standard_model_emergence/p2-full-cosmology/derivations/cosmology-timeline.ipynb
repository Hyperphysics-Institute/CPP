# Preliminary Cosmological Timeline Simulation

Simple model of lattice expansion and Capotauro breaking.

```python
import numpy as np

phi = (1 + np.sqrt(5)) / 2
N_shells = 60  # approximate e-folds
scale_factor = phi ** np.arange(0, N_shells)

print(f"Inflation scale factor after {N_shells} shells: {scale_factor[-1]:.2e}")
print(f"Equivalent e-folds: {N_shells * np.log(phi):.2f}")

# Placeholder for Capotauro time
t_capotauro = 120e6  # years
print(f"Capotauro time: {t_capotauro} years post-Big Bang")
```
