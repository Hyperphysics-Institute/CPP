# Probabilistic Strong Modes from 600-Cell Geodesics

CPP v8.0 — Samples angular layering to produce 8–10 effective gluon-like modes

**Refined breaking order**: Outer tortuous/bowed chains break first (longer path + repulsive bowing + reduced cos θ). Central chain breaks last (shortest, straightest, highest tensile strength).

```python
import numpy as np
import matplotlib.pyplot as plt
from parameters_600cell import phase_probs, phase_choices, n_events_default
```

```python
# Layer structure:
# Layer 0: 1 central path (0° — shortest, straightest)
# Layer 1: 3 paths at ~120°
# Layer 2 (third layer): variable 1–6 paths (probabilistic, mean ~4.5)
def sample_strong_modes():
    third_layer = np.random.choice(phase_choices, p=phase_probs)
    total_modes = 1 + 3 + third_layer
    return total_modes
```

### Breaking Order Refinement

Under tensile stress (quark separation):
- **Outer chains break first**: Longest/tortuous geodesics + repulsive bowing from like-charges + reduced cos θ force transmission.
- **Middle layer next**.
- **Central chain last**: Shortest path, minimal bowing, full cos θ ≈ 1 strength.

```python
# Simulate breaking energies (normalized tensile strength per layer)
def layer_breaking_strength(layer_type):
    if layer_type == 'central':
        return 1.0  # Full strength
    elif layer_type == 'middle':
        return 0.7  # ~120° paths
    else:  # outer
        return 0.4  # Tortuous + bowed

# Sequential breaking under increasing stress
def simulate_breaking_sequence(stress_levels=np.linspace(0, 2.0, 100)):
    n_modes = sample_strong_modes()
    central = 1
    middle = 3
    outer = n_modes - 4  # Remaining
    
    active_strength = []
    for stress in stress_levels:
        strength = 0
        if stress < 0.4:  # All intact
            strength = central * 1.0 + middle * 0.7 + outer * 0.4
        elif stress < 0.7:  # Outer break
            strength = central * 1.0 + middle * 0.7
        elif stress < 1.0:  # Middle break
            strength = central * 1.0
        else:  # Central breaks → confinement ends
            strength = 0
        active_strength.append(strength)
    
    return stress_levels, active_strength
```

```python
# Example breaking sequence
stress, strength = simulate_breaking_sequence()
plt.figure(figsize=(8,5))
plt.plot(stress, strength, label='Total tensile strength')
plt.axvline(0.4, color='orange', linestyle='--', label='Outer break')
plt.axvline(0.7, color='green', linestyle='--', label='Middle break')
plt.axvline(1.0, color='red', linestyle='--', label='Central break')
plt.xlabel('Normalized separation stress')
plt.ylabel('Remaining bond strength')
plt.title('Sequential Chain Breaking: Outer First, Central Last')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ensemble mode distribution (unchanged)
modes_ensemble = [sample_strong_modes() for _ in range(n_events_default)]
mean_modes = np.mean(modes_ensemble)
print(f"Mean effective modes: {mean_modes:.2f}")

plt.figure(figsize=(8,5))
plt.hist(modes_ensemble, bins=range(5,13), alpha=0.7, edgecolor='black', align='left')
plt.axvline(mean_modes, color='red', linestyle='--', label=f'Mean = {mean_modes:.2f}')
plt.axvline(8, color='green', linestyle=':', label='QCD 8 gluons')
plt.xlabel('Effective strong modes')
plt.ylabel('Frequency')
plt.title('Probabilistic Strong Modes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

This refinement strengthens confinement: outer chains fail early, releasing energy gradually until central snap triggers pair production.
