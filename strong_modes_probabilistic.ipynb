# Probabilistic Strong Modes from 600-Cell Geodesics

CPP v8.0 — Samples angular layering to produce 8–10 effective gluon-like modes

```python
import numpy as np
import matplotlib.pyplot as plt
from parameters_600cell import phase_probs, phase_choices, n_events_default
```

```python
# Layer structure: central (1) + 120° layer (3) + variable third layer
def sample_strong_modes():
    third_layer = np.random.choice(phase_choices, p=phase_probs)
    total_modes = 1 + 3 + third_layer
    return total_modes
```

```python
print(f"Example mode count: {sample_strong_modes()}")
```

```python
# Ensemble distribution
modes_ensemble = [sample_strong_modes() for _ in range(n_events_default)]
mean_modes = np.mean(modes_ensemble)
std_modes = np.std(modes_ensemble)

unique, counts = np.unique(modes_ensemble, return_counts=True)
print(f"\nEnsemble ({n_events_default:,} events):")
print(f"Mean modes: {mean_modes:.2f} ± {std_modes:.2f}")
print("Distribution:")
for u, c in zip(unique, counts):
    print(f"  {u} modes: {c/n_events_default*100:.1f}%")
```

```python
# Comparison to QCD 8 gluons
legacy_modes = 8
agreement = 100 * (1 - abs(mean_modes - legacy_modes) / legacy_modes)
print(f"\nApproximation to QCD 8 gluons: {agreement:.1f}% agreement")
```

```python
# Histogram
plt.figure(figsize=(8,5))
plt.hist(modes_ensemble, bins=range(5,13), alpha=0.7, edgecolor='black', align='left')
plt.axvline(mean_modes, color='red', linestyle='--', label=f'Mean = {mean_modes:.2f}')
plt.axvline(8, color='green', linestyle=':', label='QCD 8 gluons')
plt.xlabel('Effective strong modes')
plt.ylabel('Frequency')
plt.title('Probabilistic Strong Modes from 600-Cell Geodesics')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Asymptotic freedom proxy (high Q² suppresses layers)
def effective_modes_at_q2(q2_normalized):
    suppression = np.exp(-q2_normalized)
    return 1 + 3 * (1 - suppression) + np.random.choice(phase_choices, p=phase_probs) * suppression

q2_values = np.linspace(0, 5, 100)
eff_modes = [effective_modes_at_q2(q) for q in q2_values]

plt.figure(figsize=(8,5))
plt.plot(q2_values, eff_modes, label='Effective modes')
plt.axhline(8, color='green', linestyle=':', label='QCD baseline')
plt.xlabel('Normalized $Q^2$ (high = asymptotic)')
plt.ylabel('Active strong modes')
plt.title('Asymptotic Freedom: Mode Suppression at High $Q^2$')
plt.legend()
plt.grid(True)
plt.show()
```
