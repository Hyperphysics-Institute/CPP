# Light Hadron Spectrum (Octet/Decuplet)

CPP v8.0 — NBT + strange nesting with spin excitation

```python
import numpy as np
import matplotlib.pyplot as plt
from parameters_600cell import n_events_default
```

```python
strange_uplift_factor = 148.0  # MeV per strange
spin_excitation = 294.0       # MeV Δ-N gap
base_nucleon = 938.5          # MeV
```

```python
def baryon_mass(strangeness=0, spin=0.5, n_events=n_events_default):
    base = base_nucleon
    uplift = strangeness * strange_uplift_factor
    excitation = (spin - 0.5) * spin_excitation
    
    masses = []
    for _ in range(n_events):
        fluct = np.random.normal(1.0, 0.008)
        total = (base + uplift + excitation) * fluct
        masses.append(total)
    
    return np.mean(masses), np.std(masses)
```

```python
print("Baryon Octet (spin 1/2):")
N_mean, N_std = baryon_mass(strangeness=0, spin=0.5)
Lambda_mean, Lambda_std = baryon_mass(strangeness=1, spin=0.5)
Xi_mean, Xi_std = baryon_mass(strangeness=2, spin=0.5)

print(f"N (p/n avg): {N_mean:.1f} ± {N_std:.1f} MeV")
print(f"Λ: {Lambda_mean:.1f} ± {Lambda_std:.1f} MeV")
print(f"Ξ: {Xi_mean:.1f} ± {Xi_std:.1f} MeV")
```

```python
print("\nBaryon Decuplet (spin 3/2):")
Delta_mean, Delta_std = baryon_mass(strangeness=0, spin=1.5)
Omega_mean, Omega_std = baryon_mass(strangeness=3, spin=1.5)

print(f"Δ: {Delta_mean:.1f} ± {Delta_std:.1f} MeV")
print(f"Ω⁻: {Omega_mean:.1f} ± {Omega_std:.1f} MeV")

pdg_omega = 1672.45
agreement_omega = 100 * (1 - abs(Omega_mean - pdg_omega) / pdg_omega)
print(f"\nΩ⁻ agreement: {agreement_omega:.2f}%")
```

```python
# Plot
strange_levels = [0, 1, 2, 3]
octet_masses = [N_mean, Lambda_mean, Lambda_mean + 77, Xi_mean]  # Approx Σ split
decuplet_masses = [Delta_mean, Delta_mean + strange_uplift_factor, Delta_mean + 2*strange_uplift_factor, Omega_mean]

plt.figure(figsize=(8,5))
plt.plot(strange_levels, octet_masses, 'o-', label='Octet (spin 1/2)')
plt.plot(strange_levels, decuplet_masses, 's-', label='Decuplet (spin 3/2)')
plt.xlabel('Strangeness')
plt.ylabel('Mass (MeV)')
plt.title('Baryon Octet/Decuplet Spectrum')
plt.legend()
plt.grid(True)
plt.show()
```
