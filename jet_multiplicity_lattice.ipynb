# Jet Charged Multiplicity at √s=500 GeV

CPP v8.0 — Lattice fragmentation with probabilistic modes + tetra-core trigger

```python
import numpy as np
import matplotlib.pyplot as plt
from parameters_600cell import n_events_default
from strong_modes_probabilistic import sample_strong_modes
```

```python
base_lambda = 9.6
tetra_prob = 0.12
tetra_add_mean = 1.8
forward_boost = 0.12
```

```python
def simulate_jet_multiplicity(n_events=n_events_default):
    multiplicities = []
    for _ in range(n_events):
        base_ch = np.random.poisson(base_lambda)
        
        n_modes = sample_strong_modes()
        mode_factor = n_modes / 8.5
        
        eta = np.random.normal(0, 0.5)
        boost_ch = forward_boost * abs(eta) * base_ch * np.random.uniform(0.9, 1.1)
        
        tetra_ch = 0
        if np.random.random() < tetra_prob * mode_factor:
            tetra_ch = np.random.normal(tetra_add_mean, 0.5)
        
        total_ch = base_ch * mode_factor + boost_ch + tetra_ch + np.random.normal(0, 1.0)
        multiplicities.append(max(0, total_ch))
    
    return np.mean(multiplicities), np.std(multiplicities)
```

```python
mean_ch, std_ch = simulate_jet_multiplicity()
print(f"Jet charged multiplicity (√s=500 GeV central):")
print(f"{mean_ch:.1f} ± {std_ch:.1f}")
```

```python
exp_low, exp_high, exp_mean = 10, 13, 11.5
agreement = 100 * (1 - abs(mean_ch - exp_mean) / exp_mean) if exp_low <= mean_ch <= exp_high else 0
print(f"Agreement with experimental range: {agreement:.1f}%")
```

```python
# Histogram
multiplicities = []
for _ in range(10000):
    base_ch = np.random.poisson(base_lambda)
    n_modes = sample_strong_modes()
    mode_factor = n_modes / 8.5
    eta = np.random.normal(0, 0.5)
    boost_ch = forward_boost * abs(eta) * base_ch * np.random.uniform(0.9, 1.1)
    tetra_ch = np.random.normal(tetra_add_mean, 0.5) if np.random.random() < tetra_prob * mode_factor else 0
    total_ch = base_ch * mode_factor + boost_ch + tetra_ch + np.random.normal(0, 1.0)
    multiplicities.append(max(0, total_ch))

plt.figure(figsize=(8,5))
plt.hist(multiplicities, bins=30, alpha=0.7, edgecolor='black')
plt.axvline(mean_ch, color='red', linestyle='--', label=f'Mean = {mean_ch:.1f}')
plt.xlabel('Charged multiplicity')
plt.ylabel('Events')
plt.title('Jet Multiplicity Distribution (Lattice Fragmentation)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
