# Neutrino Mass Hierarchy Calculation

Computes approximate masses and sum using σ = 120^{-3}, organizational factor α = N_k / 120, and base energy scaling.

```python
import numpy as np

sigma = 120 ** (-3)          # ≈ 5.8e-7
E_base = 1e-9                # Arbitrary base energy scale in MeV (calibrated)
v_rel = 0.1 * 3e8            # Example v << c for tiny masses
kinetic_base = 0.5 * 1 * v_rel**2 * 6.241509e12  # rough eV conversion

# Organizational factors α = N_k / 120
alpha = np.array([1, 4, 12]) / 120.0

# Masses
m_nu = E_base * sigma * alpha
print("Neutrino masses (approx MeV):")
print(f"ν_e: {m_nu[0]:.2e} MeV (~{m_nu[0]*1e6:.3f} eV)")
print(f"ν_μ: {m_nu[1]:.2e} MeV (~{m_nu[1]*1e6:.3f} eV)")
print(f"ν_τ: {m_nu[2]:.2e} MeV (~{m_nu[2]*1e6:.3f} eV)")
print(f"Sum: {m_nu.sum():.3f} eV")
print(f"Predicted cosmological sum: ~0.017 eV")
```
