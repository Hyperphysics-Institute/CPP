# Preliminary Dark Matter Relic Density from Lattice Freeze-Out

Computes Ω_DM from freeze-out during Capotauro.

```python
import numpy as np

# Parameters
m_DM = 1000  # GeV (example)
sigma_ann = 3  # pb (weak-scale annihilation cross-section)
T_freeze = 10  # GeV (freeze-out temperature)
M_Pl = 1.22e19  # GeV

# Hubble at freeze-out
H = T_freeze**2 / M_Pl

# Relic density (standard freeze-out formula)
Omega_DM_h2 = 0.12 * (m_DM / 1000) * (3 / sigma_ann)

print(f"Freeze-out temperature: {T_freeze} GeV")
print(f"DM mass: {m_DM} GeV")
print(f"Annihilation cross-section: {sigma_ann} pb")
print(f"CPP relic density Ω_DM h²: {Omega_DM_h2:.3f}")
print(f"Observed Ω_DM h²: 0.120 (Planck 2018)")
print(f"Relative error: {abs(Omega_DM_h2 - 0.120) / 0.120 * 100:.1f}%")
```
