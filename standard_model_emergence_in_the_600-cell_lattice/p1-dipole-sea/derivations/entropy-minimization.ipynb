# Organizational Energy Demo: Random Sea vs. Polarized Cloud

Simple simulation showing entropy reduction (organizational energy) when a central CP polarizes the Dipole Sea.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N_dps = 1000           # Number of DPs in sea
r_max = 10.0           # Max radius
central_charge = -1    # -eCP

# Random sea: uniform directions
theta_random = np.random.uniform(0, 2*np.pi, N_dps)
r_random = np.random.uniform(0, r_max, N_dps)

# Polarized cloud: radial bias (simple cos theta alignment)
theta_polar = np.random.uniform(0, 2*np.pi, N_dps)
r_polar = np.random.uniform(0, r_max, N_dps)
polar_bias = np.cos(theta_polar) * (central_charge / (r_polar + 1e-3))  # Simple gradient
theta_polar += polar_bias * 0.5  # Bias directions inward

# Plot
fig, axs = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'projection': 'polar'})

axs[0].scatter(theta_random, r_random, s=5, alpha=0.5)
axs[0].set_title('Random Dipole Sea (Max Entropy)')

axs[1].scatter(theta_polar, r_polar, s=5, alpha=0.5, color='orange')
axs[1].set_title('Polarized Cloud (Organized, Lower Entropy)')

plt.tight_layout()
plt.savefig('../figures/dp-sea-random-vs-organized.png')
plt.show()

print("Note: Random sea has uniform distribution; polarized cloud shows radial alignment due to central -eCP.")
```
