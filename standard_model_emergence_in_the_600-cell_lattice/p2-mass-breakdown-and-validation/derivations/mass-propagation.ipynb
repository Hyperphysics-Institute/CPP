# Mass Propagation: From Base to PDG via Refinements

This notebook demonstrates the iterative solve for particle masses in CPP (Paper 2 Section 5). Starts with base value, adds refinements iteratively until convergence.

```python
import numpy as np

# Example: Electron (from Paper 2 table)
base = 0.306  # MeV (VEV-coupled polarization)
tolerance = 1e-6
max_iter = 20

# Simplified refinement ratios (placeholders from table)
ref_eDP_ratio = 0.33     # ~ E_eDP / m
ref_cloud_ratio = 0.10   # E_cloud
ref_residual = 0.052     # Fixed residual (SSV/spin)

m = base
delta = tolerance + 1
iteration = 0

print(f"Starting base mass: {m:.3f} MeV")

while delta > tolerance and iteration < max_iter:
    iteration += 1
    
    # Compute refinements (depend on current m)
    E_eDP = ref_eDP_ratio * m
    E_cloud = ref_cloud_ratio * m
    
    # Total refinement
    refinements = E_eDP + E_cloud + ref_residual
    
    m_new = base + refinements
    delta = abs(m_new - m)
    m = m_new
    
    print(f"Iteration {iteration}: refinements = {refinements:.3f}, new mass = {m:.3f} MeV, delta = {delta:.6f}")

print(f"\nConverged mass: {m:.3f} MeV")
print(f"PDG electron mass: 0.511 MeV")
print(f"Final error: {abs(m - 0.511):.3f} MeV")
```

## Notes
- In full model, refinements include E_inter, E_DP, σ factors, etc.
- Convergence is fast (<10 iterations) because refinements are perturbative.
- For other particles, change base and ratios accordingly (see mass-table.md).
