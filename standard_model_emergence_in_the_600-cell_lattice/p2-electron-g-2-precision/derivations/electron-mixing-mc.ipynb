# Electron Orbital ZBW Mixing Fractions (N_k=1) – Monte Carlo Refinement

Refines mixing fractions for electron (minimal cage) using simplified MC over lattice paths.

```python
import numpy as np
import matplotlib.pyplot as plt

phi = (1 + np.sqrt(5)) / 2
base_gradient = 1.0
gradients = {
    'eDP': (phi - 1) * base_gradient,  # favorable
    'hDP': base_gradient,
    'qDP': phi * base_gradient         # least favorable
}

N_k = 1  # electron
T = N_k  # thermal scale
n_samples = 10000  # MC samples

# Energy samples with small lattice noise
noise = np.random.normal(0, 0.05, n_samples)
energies = np.array([gradients['eDP'], gradients['hDP'], gradients['qDP']]) + noise[:, np.newaxis]

# Boltzmann probabilities
probs = np.exp(-energies / T)
probs /= probs.sum(axis=1)[:, np.newaxis]

# Average fractions
avg_fracs = probs.mean(axis=0)
labels = ['eDP', 'hDP', 'qDP']

print("Refined average fractions (electron, N_k=1):")
for label, frac in zip(labels, avg_fracs):
    print(f"{label}: {frac:.4f}")

# Plot distribution
plt.hist(probs[:, 0], bins=50, alpha=0.5, label='eDP')
plt.hist(probs[:, 1], bins=50, alpha=0.5, label='hDP')
plt.hist(probs[:, 2], bins=50, alpha=0.5, label='qDP')
plt.xlabel('Fraction')
plt.ylabel('Counts')
plt.title('Monte Carlo Distribution of Mixing Fractions (N_k=1)')
plt.legend()
plt.savefig('../figures/electron-mixing-histogram.png')
plt.show()
```
