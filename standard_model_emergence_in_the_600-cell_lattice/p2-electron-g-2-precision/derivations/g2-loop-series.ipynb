# Higher-Order Series for α^{-1} and Suppression S

Extends the golden-ratio series for α^{-1} and computes refined S for electron g-2.

```python
import numpy as np

phi = (1 + np.sqrt(5)) / 2

# Base series (from Appendix G)
alpha_inv_base = 360 / phi**2 - 2 / phi**3
print(f"Base α^{-1}: {alpha_inv_base:.8f}")

# Higher-order terms (from Appendix L, placeholders — refine later)
eps4 = -0.01   # example 4th order
eps5 = 0.001   # example 5th order
eps6 = -0.0001 # example 6th order

alpha_inv_higher = alpha_inv_base - eps4 / phi**4 + eps5 / phi**5 - eps6 / phi**6
print(f"Higher-order α^{-1}: {alpha_inv_higher:.8f}")

# Fine-structure constant
alpha = 1 / alpha_inv_higher
print(f"Refined α: {alpha:.10f}")

# Suppression S = α / (2π)
S = alpha / (2 * np.pi)
print(f"Refined S: {S:.6e}")

# Apply to electron mixing (from previous run)
f_qDP = 0.179519
f_hDP = 0.332917
C = 9.6e-7
mixing_sum = f_qDP + 0.7 * f_hDP
raw_delta = C * mixing_sum
delta_mu_e = raw_delta * S

print(f"Raw mixing boost: {raw_delta:.2e}")
print(f"Electron δμ (refined S): {delta_mu_e:.2e}")
```
