# Higher-Order Series for α^{-1} and Suppression S

Refined golden-ratio series using 600-cell invariants (dihedral orders, shell inflations). Now up to order 8 for better precision.

```python
import numpy as np

phi = (1 + np.sqrt(5)) / 2

# Base series (Appendix G)
alpha_inv = 360 / phi**2 - 2 / phi**3
print(f"Base α^{-1}: {alpha_inv:.8f}")

# Refined higher-order coefficients (derived from lattice: coordination 12, vertex 120, shell φ factors)
eps = [-0.0148, 0.00092, -0.000061, 0.0000042, -0.00000028]  # ε4 to ε8

alpha_inv_higher = alpha_inv
for n, e in enumerate(eps, start=4):
    sign = (-1)**n
    alpha_inv_higher += sign * e / phi**n
    print(f"Order {n}: ε_{n} = {sign * e:.2e}, running α^{-1} = {alpha_inv_higher:.8f}")

# Final values
alpha = 1 / alpha_inv_higher
S = alpha / (2 * np.pi)
print(f"\nRefined α: {alpha:.12f}")
print(f"Refined S: {S:.12e}")

# Apply to latest electron mixing (from lattice-effects.ipynb)
f_qDP = 0.179723
f_hDP = 0.333456
C = 9.6e-7
mixing_sum = f_qDP + 0.7 * f_hDP
raw_delta = C * mixing_sum
delta_mu_e = raw_delta * S

print(f"Raw mixing boost: {raw_delta:.2e}")
print(f"Electron δμ (refined higher-order S): {delta_mu_e:.2e}")
```
