# Monte Carlo for DP Cloud Averages

Simulates radial distributions for quarks (gradients) vs. leptons (uniform).

```python
import numpy as np
import matplotlib.pyplot as plt

# Quark example: Radial Boltzmann
r = np.linspace(0, 1, 100)  # Normalized radius
types = ['qDP', 'hDP', 'eDP', 'AB']
ssv_ratios = [4, 2.5, 1, 1.5]  # Example ratios
P = np.exp(-np.outer(r, ssv_ratios))  # Simplified exp(-γ r * SSV)
P /= P.sum(axis=1)[:, np.newaxis]

plt.stackplot(r, P.T, labels=types)
plt.xlabel('Normalized Radius')
plt.ylabel('Fraction')
plt.title('Quark Radial DP Gradient')
plt.legend()
plt.savefig('../figures/radial-gradient-plot.pdf')
plt.show()
```
