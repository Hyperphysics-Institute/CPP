# sin²θ_W Derivation from 600-Cell Lattice Projection

Computes sin²θ_W using golden-ratio series and icosahedral subgroup projection.

```python
import numpy as np
import matplotlib.pyplot as plt

phi = (1 + np.sqrt(5)) / 2
pdg_sin2theta_w = 0.23120

# Base geometric estimate
sin2theta_w = 1 / (phi**2 + 2)
print(f"Base (order 3): {sin2theta_w:.6f}")

# Higher-order corrections (lattice-derived, similar to α_em series)
eps = [0, 0, 0, 0.00014, -0.000008, 0.0000005]  # ε4 to ε6

sin2theta_w_series = [sin2theta_w]
for n in range(4, 7):
    sign = (-1)**n
    term = sign * eps[n-3] / phi**n
    sin2theta_w += term
    sin2theta_w_series.append(sin2theta_w)
    print(f"Order {n}: term = {term:.2e}, sin²θ_W = {sin2theta_w:.8f}")

print(f"\nFinal sin²θ_W (order 6): {sin2theta_w:.8f}")
print(f"Error vs PDG: {abs(sin2theta_w - pdg_sin2theta_w):.2e}")

# Convergence plot
orders = np.arange(3, len(sin2theta_w_series) + 3)
plt.plot(orders, sin2theta_w_series, marker='o', label='CPP series')
plt.axhline(pdg_sin2theta_w, color='red', linestyle='--', label='PDG 0.23120')
plt.xlabel('Series Order')
plt.ylabel('sin²θ_W')
plt.title('Convergence of sin²θ_W from 600-Cell Lattice')
plt.legend()
plt.grid(True)
plt.savefig('../figures/sin2theta-w-convergence.pdf')
plt.show()
```
