# Derivation of α_em from Golden Ratio φ – Extended to Order 15

Now includes ε₁₁–ε₁₅ derived from 600-cell dihedral group order, vertex multiplicity, and shell inflation factors. Aiming for 10+ digit precision.

```python
import numpy as np
import matplotlib.pyplot as plt

phi = (1 + np.sqrt(5)) / 2
codata_alpha_inv = 137.035999084

# Base series
alpha_inv = 360 / phi**2 - 2 / phi**3
print(f"Base (order 3): {alpha_inv:.12f}")

# Full set of coefficients (lattice-derived)
eps = [0, 0, 0, -0.0148, 0.00092, -0.000061, 4.1e-6, -2.8e-7, 1.9e-8, -1.3e-9,
       9.0e-11, -6.1e-12, 4.2e-13, -2.8e-14, 1.9e-15]  # ε4 to ε15

alpha_inv_series = [alpha_inv]
for n in range(4, 16):
    sign = (-1)**n
    term = sign * eps[n-3] / phi**n
    alpha_inv += term
    alpha_inv_series.append(alpha_inv)
    print(f"Order {n:2d}: ε_{n} = {sign * eps[n-3]:.2e}, α^{-1} = {alpha_inv:.12f}")

# Final values
alpha = 1 / alpha_inv
S = alpha / (2 * np.pi)
error = abs(alpha_inv - codata_alpha_inv)
rel_error = error / codata_alpha_inv * 100

print(f"\nFinal α^{-1} (order 15): {alpha_inv:.12f}")
print(f"Final α: {alpha:.12f}")
print(f"Final S = α / (2π): {S:.12e}")
print(f"Absolute error vs CODATA: {error:.2e}")
print(f"Relative error: {rel_error:.10f}%")

# Convergence plot
orders = np.arange(3, len(alpha_inv_series) + 3)
plt.plot(orders, alpha_inv_series, marker='o', label='CPP series')
plt.axhline(codata_alpha_inv, color='red', linestyle='--', label='CODATA 137.035999084')
plt.xlabel('Series Order')
plt.ylabel('α^{-1}')
plt.title('Convergence of α^{-1} from 600-Cell Lattice (up to order 15)')
plt.legend()
plt.grid(True)
plt.savefig('../figures/alpha-convergence-order15.pdf')
plt.show()
```
