# Derivation of α_em from Golden Ratio φ

Computes α^{-1} series and convergence to CODATA value.

```python
import numpy as np
import matplotlib.pyplot as plt

phi = (1 + np.sqrt(5)) / 2
codata_alpha_inv = 137.035999084

# Base series
alpha_inv = 360 / phi**2 - 2 / phi**3
print(f"Base (order 3): {alpha_inv:.8f}")

# Higher-order coefficients (lattice-derived)
eps = [0, 0, 0, -0.0148, 0.00092, -0.000061]  # index 3=base, 4-6=ε4-ε6

alpha_inv_series = [alpha_inv]
for n in range(4, 10):
    sign = (-1)**n
    term = sign * eps[n-3] / phi**n if n-3 < len(eps) else 0
    alpha_inv += term
    alpha_inv_series.append(alpha_inv)
    print(f"Order {n}: term = {term:.2e}, α^{-1} = {alpha_inv:.8f}")

# Plot convergence
orders = np.arange(3, len(alpha_inv_series) + 3)
plt.plot(orders, alpha_inv_series, marker='o', label='CPP series')
plt.axhline(codata_alpha_inv, color='red', linestyle='--', label='CODATA')
plt.xlabel('Series Order')
plt.ylabel('α^{-1}')
plt.title('Convergence of α^{-1} from φ Series')
plt.legend()
plt.grid(True)
plt.savefig('../figures/alpha-convergence.pdf')
plt.show()

print(f"Final α^{-1}: {alpha_inv:.12f}")
print(f"Error vs CODATA: {abs(alpha_inv - codata_alpha_inv):.2e}")
```
