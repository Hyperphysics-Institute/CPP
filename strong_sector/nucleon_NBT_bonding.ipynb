# Nucleon Binding Tetra (NBT) Bonding

CPP v8.0 — Primary compact bonding + stress-dependent chain insertion

```python
import numpy as np
import matplotlib.pyplot as plt
from parameters_600cell import n_events_default
```

```python
def nbt_bonding_energy(stress_level=0.0, use_variation=False):
    base = 1.0
    chains_inserted = stress_level**2
    qdp_contrib = min(chains_inserted, 1.0) * 1.2
    hdp_edp_contrib = max(0, chains_inserted - 1.0) * 0.8
    total = base + qdp_contrib + hdp_edp_contrib
    if use_variation:
        total += np.random.normal(0, 0.02)
    return total
```

```python
def nucleon_mass(proton=True, stress_level=0.0):
    base_nucleon = 938.0
    bonding = nbt_bonding_energy(stress_level)
    polarity = 0.15 if proton else -0.10
    return base_nucleon * (1 + polarity * 0.001 + (bonding - 1.0) * 0.01)
```

```python
proton_mass = nucleon_mass(proton=True)
neutron_mass = nucleon_mass(proton=False)
print(f"Unstressed NBT:")
print(f"Proton mass: {proton_mass:.1f} MeV")
print(f"Neutron mass: {neutron_mass:.1f} MeV")
```

```python
stress_levels = np.linspace(0, 1.0, 10)
proton_stressed = [nucleon_mass(proton=True, stress_level=s) for s in stress_levels]
neutron_stressed = [nucleon_mass(proton=False, stress_level=s) for s in stress_levels]

plt.figure(figsize=(8,5))
plt.plot(stress_levels, proton_stressed, 'o-', label='Proton (uud)')
plt.plot(stress_levels, neutron_stressed, 's-', label='Neutron (udd)')
plt.xlabel('Stress level (normalized)')
plt.ylabel('Effective mass / binding energy (MeV)')
plt.title('NBT Primary Bonding + Stress-Dependent Chain Strengthening')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mean_proton, std_proton = np.mean([nucleon_mass(proton=True, stress_level=0.3, use_variation=True) 
                                   for _ in range(n_events_default)]), \
                         np.std([nucleon_mass(proton=True, stress_level=0.3, use_variation=True) 
                                 for _ in range(n_events_default)])

print(f"\nEnsemble (moderate stress): Proton {mean_proton:.1f} ± {std_proton:.1f} MeV")
```
