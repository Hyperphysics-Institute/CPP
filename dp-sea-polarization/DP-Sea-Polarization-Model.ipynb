## Velocity-Dependent DP Sea Polarization & Inertial Resistance

Inspired by dream insight after 60-probe handedness paper.

Goal: reproduce silly-putty behavior (transparent → viscous → solid) and converge to Lorentz γ(v).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c

# Constants (tunable)
v0 = c          # characteristic velocity (m/s)
alpha = 1.8     # sharpness of transition
k_omega = 2.0   # frequency coupling
beta = 3.0      # closest-approach coupling
k_q = 1.5       # charge enhancement
k_f = 1.0       # force multiplier
omega0 = 1e20   # base ZBW frequency (rad/s)
d0 = 1e-15      # base closest approach (m)

def P(v, r=None):
    """Polarization of DP Sea (v in m/s)"""
    if r is None:
        return np.tanh(v / v0)
    else:
        v0_r = v0 * (r / 1e-10)**0.5  # example radial dependence
        return np.tanh(v / v0_r)

# Example plot
v = np.linspace(0, 0.999*c, 1000)
plt.figure(figsize=(10,6))
plt.plot(v/c, P(v), label='P(v) (tanh)')
plt.xlabel('v / c')
plt.ylabel('Polarization P')
plt.title('DP Sea Polarization vs Velocity')
plt.grid(True)
plt.legend()
plt.show()
```

## Tranche 2 – Oscillation Parameters

```python
def omega(v):
    return omega0 * np.exp(k_omega * P(v))

def d_min(v):
    return d0 * np.exp(-beta * P(v))

plt.figure(figsize=(10,6))
plt.semilogy(v/c, omega(v)/omega0, label='ω(v)/ω₀')
plt.semilogy(v/c, d_min(v)/d0, label='d_min(v)/d₀')
plt.xlabel('v / c')
plt.ylabel('Relative value')
plt.title('Oscillation Parameters vs Velocity')
plt.grid(True)
plt.legend()
plt.show()
```

## Tranche 3 – Repulsive Force (at closest approach)

```python
def F_rep(v):
    return (1 + k_q * P(v))**2 * (omega(v)/omega0) * (d0 / d_min(v))**2

plt.figure(figsize=(10,6))
plt.semilogy(v/c, F_rep(v), label='F_rep(v) / F₀')
plt.xlabel('v / c')
plt.ylabel('Relative force amplification')
plt.title('Repulsive Force Amplification vs Velocity')
plt.grid(True)
plt.legend()
plt.show()
```

## Tranche 4 – Effective Mass & Lorentz Convergence

```python
def gamma(v):
    return 1 / np.sqrt(1 - (v/c)**2)

def m_eff(v, k=0.04):
    # Simple linear + gamma term
    return 1 + k * (v/c) * gamma(v)

plt.figure(figsize=(10,6))
plt.plot(v/c, gamma(v), 'k--', label='Lorentz γ(v)')
plt.plot(v/c, m_eff(v), label='m_eff(v) / m₀')
plt.xlabel('v / c')
plt.ylabel('Relative mass')
plt.title('Effective Mass vs Velocity')
plt.grid(True)
plt.legend()
plt.show()
```
