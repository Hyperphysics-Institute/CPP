# Nucleon Magnetic Moments from ZBW Currents

CPP v8.0 — NBT topology + polarity bias

```python
import numpy as np
import matplotlib.pyplot as plt
from parameters_600cell import n_events_default
```

```python
base_g = 2.0

def magnetic_moment(proton=True, use_variation=False):
    polarity_bias = 0.15 if proton else -0.10
    anomaly_base = 0.792
    anomaly = anomaly_base + polarity_bias
    sign = 1 if proton else -1
    suppression = 1.0 if proton else 0.98
    total = sign * (base_g / 2 + anomaly * suppression)
    if use_variation:
        total += np.random.normal(0, 0.001)
    return total
```

```python
proton_mu = magnetic_moment(proton=True)
neutron_mu = magnetic_moment(proton=False)
print(f"Proton μ: {proton_mu:+.4f} µ_N")
print(f"Neutron μ: {neutron_mu:+.4f} µ_N")
```

```python
proton_ensemble = [magnetic_moment(proton=True, use_variation=True) for _ in range(n_events_default)]
neutron_ensemble = [magnetic_moment(proton=False, use_variation=True) for _ in range(n_events_default)]

proton_mean = np.mean(proton_ensemble)
proton_std = np.std(proton_ensemble)
neutron_mean = np.mean(neutron_ensemble)
neutron_std = np.std(neutron_ensemble)

print(f"\nEnsemble ({n_events_default:,} events):")
print(f"Proton: {proton_mean:+.4f} ± {proton_std:.4f} µ_N")
print(f"Neutron: {neutron_mean:+.4f} ± {neutron_std:.4f} µ_N")
```

```python
pdg_proton = 2.792847
pdg_neutron = -1.913043

agreement_p = 100 * (1 - abs(proton_mean - pdg_proton) / pdg_proton)
agreement_n = 100 * (1 - abs(neutron_mean - pdg_neutron) / abs(pdg_neutron))

print(f"\nAgreement:")
print(f"Proton: {agreement_p:.2f}%")
print(f"Neutron: {agreement_n:.2f}%")
```

```python
plt.figure(figsize=(8,5))
plt.hist(proton_ensemble, bins=50, alpha=0.6, label='Proton', color='blue')
plt.hist(neutron_ensemble, bins=50, alpha=0.6, label='Neutron', color='red')
plt.axvline(pdg_proton, color='blue', linestyle='--')
plt.axvline(pdg_neutron, color='red', linestyle='--')
plt.xlabel('Magnetic moment (µ_N)')
plt.ylabel('Events')
plt.title('ZBW-Derived Nucleon Magnetic Moments')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
