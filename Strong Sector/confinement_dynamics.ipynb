# Confinement Dynamics in CPP
Simulates string breaking and fraying with stress thresholds using hDP insights.

```python
import numpy as np

# Parameters
stress_threshold = 1.5  # Arbitrary units for fraying start
sea_pair_prob = 0.8  # Probability of sea qDP bonding
n_steps = 100  # Stretching steps

def simulate_fraying(initial_length=1.0, stretch_rate=0.05):
    length = initial_length
    force = 0
    broke = False
    for step in range(n_steps):
        length += stretch_rate
        force = length  # Linear confinement
        if force > stress_threshold:
            if np.random.random() < sea_pair_prob:
                broke = True
                break
    return broke, force, length

# Run ensemble
n_events = 1000
breaks = [simulate_fraying() for _ in range(n_events)]
break_rate = sum(b[0] for b in breaks) / n_events
print(f'Break rate: {break_rate:.2f}')
print(f'Avg final force: {np.mean([b[1] for b in breaks]):.2f}')
print(f'Avg break length: {np.mean([b[2] for b in [bb for bb in breaks if bb[0]]):.2f}')
```
